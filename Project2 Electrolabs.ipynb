```python
import seaborn as sns
df.describe()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x= df.Date
y= df.Pen
z= df.Pencil
plt.plot(x,y,'r*-')
plt.xlabel('Date')
plt.ylabel('Number')
plt.title('Sales of Pens')
```

```python
plt.plot(x,z,'g^-')
plt.xlabel('Date')
plt.ylabel('Number')
plt.title('Sales of Pencils')
```

```python
plt.plot(x,z,'g')
plt.plot(x,y,'r--')
plt.xlabel('Date')
plt.ylabel('Number')
plt.title('Sales combined')
```
